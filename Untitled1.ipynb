```python
import pygraphviz as pgv
import matplotlib.pyplot as plt
from PIL import Image

G = pgv.AGraph(driected=False)
G.node_attr['shape'] = 'circle'
G.edge_attr['dir'] = 'forward'
G.edge_attr['style'] = 'solid'

G.graph_attr['dpi'] = 200
G.graph_attr['rankdir'] = "LR"
G.graph_attr['splines'] = False
G.graph_attr['ranksep'] = 1.4


for i in range(10):
    for j in range(5):
        G.add_edge(f"a{i}", f"b{j}", color="#66ccff", penwidth=0.7)


G.layout("dot")
G.draw("test.png")
plt.figure(figsize=(7, 15))
plt.axis('off')
plt.imshow(Image.open("test.png"))
plt.show()
```
