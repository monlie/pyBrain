```python
import pygraphviz as pgv
import matplotlib.pyplot as plt
from PIL import Image

G = pgv.AGraph(driected=False)
G.node_attr['shape'] = 'circle'
G.edge_attr['dir'] = 'forward'
G.edge_attr['style'] = 'solid'

G.graph_attr['dpi'] = 200
G.graph_attr['rankdir'] = "RL"
G.graph_attr['splines'] = False
G.graph_attr['ranksep'] = 1.4

G.add_node('A')
G.add_edge('c', 'd', color="blue", penwidth=2)
G.add_edge('d', 'm', color="blue", penwidth=2)


G.layout()
G.draw("test.png")
plt.axis('off')
plt.imshow(Image.open("test.png"))
plt.show()
```
