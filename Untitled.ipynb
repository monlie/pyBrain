```python
import numpy as np
import matplotlib.pyplot as plt


class IzhikevichNeuron(object):
    
    def __init__(self, ts=5e-2, a=0.02, b=0.2, c=-65.0, d=8.0):
        self.time_stamp = None
        self.status = np.array([-70.0, -14.0], dtype=np.float64)
        self.time_step = ts
        
        self._a = a
        self._b = b
        self._c = c
        self._d = d

    def izhi_equs(self, x, current):
        v, u = x
        dv = 0.04 * v*v + 5 * v + 140 - u + current
        du = self._a * (self._b * v - u)
        return np.array([dv, du], dtype=np.float64)
    
    def reset(self):
        self.status[0] = self._c
        self.status[1] += self._d
        
    def get_current(self, t):
        return 30 if t > 25 else 0
    
    # integrate using rk45
    def one_step_update(self, x, t, dt):
        halfdt = dt * 0.5
        k1 = self.izhi_equs(x, self.get_current(t))
        k2 = self.izhi_equs(x + halfdt * k1, self.get_current(t + halfdt))
        k3 = self.izhi_equs(x + halfdt * k2, self.get_current(t + halfdt))
        k4 = self.izhi_equs(x + dt * k3, self.get_current(t + dt))
        x = x + (k1 + 2 * k2 + 2 * k3 + k4) * dt / 6
        return x
    
    def update_status(self, steps=4000):
        t = 0.0
        timing = [t]
        solution = [self.status[0]]
        spiking = []
        for i in range(steps):
            dt = self.time_step
            x = self.one_step_update(self.status, t, dt)
            if x[0] >= 30:
                # binary search the precise time of spiking
                x, dt = self.search_t(t, 0, dt)
                self.reset()
                spiking.append(t + dt)
            else:
                self.status = x
            solution.append(x[0])
            t += dt
            timing.append(t)
        print(self.status)
        return timing, solution, spiking
    
    def fire(self):
        pass
    
    def search_t(self, t, ldt, rdt):
        dt = (rdt + ldt) * 0.5
        x = self.one_step_update(self.status, t, dt)
        if x[0] >= 30:
            return self.search_t(t, ldt, dt)
        if 30 - x[0] > 1e-4:
            return self.search_t(t, dt, rdt)
        return x, dt
```

```python
izhi = IzhikevichNeuron(d=8)
timing, solution, spiking = izhi.update_status()
plt.figure(figsize=(10, 3))

plt.plot(timing, solution)
plt.scatter(spiking, [30]*len(spiking), color="orange")
plt.show()
```
