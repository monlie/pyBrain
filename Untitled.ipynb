```python
import numpy as np
import matplotlib.pyplot as plt
import neuron
import oscilloscope
import synapse
from random import random
import integrator
    
    
def possion_spiking(neuron, t, dt):
    p = 0.05
    cond = random() <= dt * p
    if cond:
        for syn in neuron.presynapses:
            syn.simulate(t)
    return cond
```

```python
test_neuron = neuron.NeuronFactory("RS", lambda t: (0 if t < 100 or t > 700 else -10))
osi = oscilloscope.VoltageOscilloscope(test_neuron)
itr = integrator.Integrator(test_neuron, 0.1)
itr.integrate(8000, callback=None)

#draw
fig = plt.figure(figsize=(14, 3))
osi.display(fig, lw=1.5)
plt.show()
```

```python
test_neuron = neuron.NeuronFactory("RS")
test_syn = synapse.ChemicalSynapse(None, 0.3, 0.4)
test_syn.delay = 0
test_neuron.presynapses.append(test_syn)
osi = oscilloscope.VoltageOscilloscope(test_neuron)
osi2 = oscilloscope.SpikingOscilloscope(test_syn)

itr = integrator.Integrator(test_neuron, 0.1)
itr.integrate(8000, callback=possion_spiking)
fig = plt.figure(figsize=(14, 3))
osi.display(fig, lw=1.5)
osi2.display(fig, s=7, color="red")
plt.show()
```

```python
%matplotlib inline
test_neuron = neuron._TestNeuron(np.array([1, 0], dtype=np.float64))
osi = oscilloscope.VoltageOscilloscope(test_neuron)

itr = integrator.Integrator(test_neuron, 0.01)
itr.integrate(600)
fig = plt.figure(figsize=(14, 3))

osi.display(fig, lw=1.8)
osi.reboot()
test_neuron.status[0] = 1

itr = integrator.EulerIntegrator(test_neuron, 0.01)
itr.integrate(600)
osi.display(fig, lw=1.8)
plt.show()
```
