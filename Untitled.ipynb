```python
import numpy as np
import matplotlib.pyplot as plt
import neuron
import oscilloscope
import synapse
from random import random
    
    
def possion_spiking(neuron, t, dt):
    p = 1
    cond = random() <= dt * p
    if cond:
        for syn in neuron.presynapses:
            syn.simulate(t)
    return cond
```

```python
test_neuron = neuron.IzhikevichNeuron(d=8)
test_neuron.presynapses.append(synapse.ChemicalSynapse(None, 0.3, 0.2))
osi = oscilloscope.VoltageOscilloscope(test_neuron)
# osi2 = oscilloscope.PreSpikingOscilloscope(test_neuron)
test_neuron.integrate(2000, callback=possion_spiking)
fig = plt.figure(figsize=(14, 3))
osi.display(fig, lw=1.8)
# osi2.display(fig, s=7, color="red")
plt.show()
```
