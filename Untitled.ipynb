```python
import numpy as np
import matplotlib.pyplot as plt
import neuron
import oscilloscope
import synapse
from random import random
import integrator
    
    
def possion_spiking(neuron, t, dt):
    p = 0.9
    cond = random() <= dt * p
    if cond:
        for syn in neuron.presynapses:
            syn.simulate(t)
    return cond
```

```python
test_neuron = neuron.IzhikevichNeuron(d=8)
test_syn = synapse.ChemicalSynapse(None, 0.3, 0.4)
test_neuron.presynapses.append(test_syn)
osi = oscilloscope.VoltageOscilloscope(test_neuron)
osi2 = oscilloscope.SpikingOscilloscope(test_syn)

itr = integrator.Integrator(test_neuron)
itr.integrate(2000, callback=possion_spiking)
fig = plt.figure(figsize=(14, 3))
osi.display(fig, lw=1.8)
osi2.display(fig, s=7, color="red")
plt.show()
```
