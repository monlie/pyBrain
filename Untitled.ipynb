```python
import numpy as np
import matplotlib.pyplot as plt
from Neurons import neuron
from Oscilloscopes import oscilloscope
from Synapses import synapse
from random import random
from Integrators import integrator
    
    
def possion_spiking(neuron, t, dt):
    p = 0.08
    cond = random() <= dt * p
    if cond:
        for syn in neuron.presynapses:
            syn.simulate(t)
    return cond
```

```python
def get_votage(neuron_type, ax):
    test_neuron = neuron.NeuronFactory(neuron_type, lambda t: (0 if t < 50 or t > 350 else -10))
    osi = oscilloscope.VoltageOscilloscope(test_neuron)
    itr = integrator.Integrator(test_neuron, 0.1)
    itr.integrate(4000, callback=None)
    
    #draw
    osi.display(ax, lw=1.2)
    

fig, axs = plt.subplots(3, 1, figsize=(8, 8), sharex="col")
for nt, ax in zip(["RS", "IB", "CH"], axs):
    get_votage(nt, ax)
    ax.set_ylabel(r"$V\ [mV]$")
    ax.annotate(f"type {nt}", (0.89, 0.85), xycoords='axes fraction')
ax.set_xlabel(r"$t\ [ms]$")
plt.savefig("izhineurons.pdf", bbox_inches='tight', pad_inches=0.0)
plt.show()
```

```python
test_neuron = neuron.NeuronFactory("RS")
test_syn = synapse.ChemicalSynapse(None, 0.25, 0.4, 20)
test_neuron.presynapses.append(test_syn)
osi = oscilloscope.VoltageOscilloscope(test_neuron)
osi2 = oscilloscope.SpikingOscilloscope(test_syn)

itr = integrator.Integrator(test_neuron, 0.1)
itr.integrate(4000, callback=possion_spiking)
fig = plt.figure(figsize=(14, 3))
osi.display(fig, lw=1.5)
osi2.display(fig, s=7, color="red")
plt.show()
```
